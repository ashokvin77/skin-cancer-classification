```python
import pandas as pd
from pathlib import Path

# Same project root as notebook 1. The DATA folder already has the CSV.
PROJECT = Path('/content/drive/MyDrive/derm9')
DATA    = PROJECT / 'data'

df = pd.read_csv(DATA / 'derm9_curated.csv')

print(f'Rows: {len(df):,}')
df.sample(7)
```

## Pool and stratified split (60/20/20)
**Goal.** Build three DataFrames — `train_df`, `val_df`, `test_df` with class proportions preserved across all three splits.

**Why pool and re-split.** The original `Test/` folder ships with only 3 images for two of the rare classes. Too thin to evaluate per-class performance reliably.

```python
from sklearn.model_selection import train_test_split

SEED = 99

# 20% test, 80% trainval (stratified by class).
trainval_df, test_df = train_test_split(
    df,
    test_size=0.20,
    stratify=df['class_orig'],
    random_state=SEED,
)

# Second split
train_df, val_df = train_test_split(
    trainval_df,
    test_size=0.25,
    stratify=trainval_df['class_orig'],
    random_state=SEED,
)

# Reset indices so each DataFrame has clean 0..N-1 indexing.
train_df = train_df.reset_index(drop=True)
val_df   = val_df.reset_index(drop=True)
test_df  = test_df.reset_index(drop=True)

total = len(df)
print(f'Train: {len(train_df):>5,}  ({len(train_df) / total * 100:.1f}%)')
print(f'Val  : {len(val_df):>5,}  ({len(val_df)  / total * 100:.1f}%)')
print(f'Test : {len(test_df):>5,}  ({len(test_df) / total * 100:.1f}%)')
```

## Verify the split

Confirm no image appears in more than one split.



```python
def check_split_overlap(train_df, val_df, test_df):
    """Print overlap findings between splits. Returns True if any pair overlaps."""
    train_imgs = set(train_df['image'])
    val_imgs   = set(val_df['image'])
    test_imgs  = set(test_df['image'])

    pairs = {
        'train and val':  train_imgs & val_imgs,
        'train and test': train_imgs & test_imgs,
        'val and test':   val_imgs   & test_imgs,
    }

    any_leak = False
    for label, overlap in pairs.items():
        if overlap:
            any_leak = True
            print(f'LEAK: {label} share {len(overlap):,} filenames')
        else:
            print(f'OK  : {label} disjoint')

    return any_leak


overlap_found = check_split_overlap(train_df, val_df, test_df)
```

All three pairs overlap. That shouldn't happen. Time to
investigate.

## Diagnose the overlap

Figure out where the apparent leak is coming from.



```python
# How many filenames in df appear in more than one row?
counts = df['image'].value_counts()
duplicates = counts[counts > 1]

print(f'Filenames appearing in >1 row of df: {len(duplicates):,}')

# Look at one example to see what the conflict looks like.
example = duplicates.index[0]
print(f'\nExample: {example}')
df[df['image'] == example][['class_orig', 'split_orig', 'filepath']]
```

```python
# Confirm visually: are the duplicated filenames actually the same image?

import matplotlib.pyplot as plt
from PIL import Image

# Look at the first 3 duplicate filenames.
sample_names = duplicates.index[:3]

fig, axes = plt.subplots(len(sample_names), 2, figsize=(8, 4 * len(sample_names)))

for row, name in enumerate(sample_names):
    pair = df[df['image'] == name].reset_index(drop=True)
    for col, (_, sample) in enumerate(pair.iterrows()):
        ax = axes[row, col]
        ax.imshow(Image.open(sample['filepath']))
        ax.axis('off')
        ax.set_title(f"{sample['class_orig']}\n{name}", fontsize=9)

plt.tight_layout()
plt.show()
```

The image content is identical across both copies, only the class label differs.

```python
# Do all 155 duplicate filenames have identical file bytes?
import hashlib
def file_hash(path):
    """MD5 of the file bytes. Identical bytes give identical hash."""
    with open(path, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()


# For each duplicate filename, do all its copies have the same bytes?
identical = different = 0
for name in duplicates.index:
    paths = df.loc[df['image'] == name, 'filepath'].tolist()
    if len({file_hash(p) for p in paths}) == 1:
        identical += 1
    else:
        different += 1

print(f'Identical bytes (curation error) : {identical}')
print(f'Different bytes (filename clash) : {different}')
```

All 155 duplicate filenames have byte-identical content. The dataset really does contain the same image filed under two different diagnoses; a curation error in the source. Removing both copies of each pair is the right call.

## Remove duplicate-label conflicts

Drop both copies of every ambiguous image.

**Why drop both, not pick one.** We don't know which label is correct.


```python
# Set of filenames that appear in >1 row.
ambiguous = set(duplicates.index)

n_before = len(df)
df = df[~df['image'].isin(ambiguous)].reset_index(drop=True)
n_after = len(df)

print(f'Rows before: {n_before:,}')
print(f'Rows after : {n_after:,}')
print(f'Removed    : {n_before - n_after:,}  ({(n_before - n_after) / n_before * 100:.1f}%)')

# Filenames should now be unique.
assert df['image'].nunique() == len(df), 'Filenames still not unique'
print('Every filename is now unique.')
```

```python
trainval_df, test_df = train_test_split(
    df,
    test_size=0.20,
    stratify=df['class_orig'],
    random_state=SEED,
)
train_df, val_df = train_test_split(
    trainval_df,
    test_size=0.25,
    stratify=trainval_df['class_orig'],
    random_state=SEED,
)

train_df = train_df.reset_index(drop=True)
val_df   = val_df.reset_index(drop=True)
test_df  = test_df.reset_index(drop=True)

total = len(df)
print(f'Train: {len(train_df):>5,}  ({len(train_df) / total * 100:.1f}%)')
print(f'Val  : {len(val_df):>5,}  ({len(val_df)  / total * 100:.1f}%)')
print(f'Test : {len(test_df):>5,}  ({len(test_df) / total * 100:.1f}%)')
```

## Define transforms

Define the image preprocessing pipelines (resize, crop, optional augment, convert to tensor) that each image will pass through before reaching the model.

One without augmentation (used by the baseline model and by all eval). One with augmentation (used by the augmented model's training).

```python
import torch
from torchvision.transforms import v2

# No augmentation: resize, crop, to tensor in [0, 1]. Used by the baseline
# model and by all evaluation.
basic_tfm = v2.Compose([
    v2.Resize(256, antialias=True),
    v2.CenterCrop(224),
    v2.PILToTensor(),
    v2.ToDtype(torch.float32, scale=True),
])

# With augmentation: same as basic plus random flips, rotation, color jitter.
# Used only for training the augmented model.
augmented_tfm = v2.Compose([
    v2.Resize(256, antialias=True),
    v2.CenterCrop(224),
    v2.RandomHorizontalFlip(p=0.5),
    v2.RandomVerticalFlip(p=0.5),
    v2.RandomRotation(15),
    v2.ColorJitter(0.1, 0.1, 0.1, 0.02),
    v2.PILToTensor(),
    v2.ToDtype(torch.float32, scale=True),
])

print('Transforms ready: basic_tfm, augmented_tfm')
```

## PyTorch Dataset class

Tell PyTorch how to read one image and its label from our DataFrame.

PyTorch needs a small object that knows two things: how many images we have, and how to fetch the i-th one. We write that object once and reuse it for train, val, and test.

```python
from torch.utils.data import Dataset
from PIL import Image

class SkinLesionDataset(Dataset):
    """One row of the DataFrame becomes one (image_tensor, label) sample."""

    def __init__(self, df, class_to_idx, transform=None):
        # reset_index ensures self.df.iloc[i] works regardless of how the
        # incoming DataFrame was sliced.
        self.df = df.reset_index(drop=True)
        self.class_to_idx = class_to_idx
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # .convert('RGB') guarantees 3 channels even if a source file is
        # grayscale or has an alpha channel.
        img = Image.open(row['filepath']).convert('RGB')

        if self.transform is not None:
            img = self.transform(img)

        label = self.class_to_idx[row['class_orig']]
        return img, label
```

## Instantiate Datasets and DataLoaders
Create three datasets (train, val, test) and wrap each in a `DataLoader` that yields batches.



```python
from torch.utils.data import DataLoader

BATCH_SIZE = 32

# Three Datasets — train uses augmented_tfm, val and test use basic_tfm.
train_ds = SkinLesionDataset(train_df, class_to_idx, transform=augmented_tfm)
val_ds   = SkinLesionDataset(val_df,   class_to_idx, transform=basic_tfm)
test_ds  = SkinLesionDataset(test_df,  class_to_idx, transform=basic_tfm)

# Three DataLoaders — train shuffles, val and test don't.
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE,
                          shuffle=True,
                          num_workers=2)
val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE,
                          shuffle=False,
                          num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE,
                          shuffle=False,
                          num_workers=2)

print(f'Train: {len(train_ds):,} samples, {len(train_loader):,} batches')
print(f'Val  : {len(val_ds):,} samples, {len(val_loader):,} batches')
print(f'Test : {len(test_ds):,} samples, {len(test_loader):,} batches')
```

## Smoke test

Pull one training batch through the pipeline end-to-end and confirm everything looks right.


**What we check.**
- **Shape.** Should be `(32, 3, 224, 224)` — 32 images per batch, 3 channels (RGB), 224×224 spatial.
- **Dtype.** Pixels should be `torch.float32`, labels should be `torch.int64`.
- **Pixel value range.** Should sit inside `[0, 1]` because the transform pipeline scales pixels that way.
- **Label diversity.** A shuffled batch should contain several classes, not all 32 samples from one class.

```python
# Pull one batch from the train loader. iter() turns the loader into a
# Python iterator; next() asks for the first batch.
images, labels = next(iter(train_loader))

print(f'Image batch shape : {tuple(images.shape)}')
print(f'Label batch shape : {tuple(labels.shape)}')
print(f'Image dtype       : {images.dtype}')
print(f'Label dtype       : {labels.dtype}')
print(f'Pixel value range : [{images.min():.3f}, {images.max():.3f}]')
print(f'Unique labels     : {labels.unique().tolist()}')
```
